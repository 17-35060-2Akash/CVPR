```python
import numpy as np
b=np.array([7,2,9,1,5])
print("::Bubble Sort:: \n")

print("The Unsorted List is:",b)
print("\n")
for j in range(len(b)-1): 
  print(f"----iteration: {j+1}----")
  for i in range(len(b)-1):
    print(f"--Comparison: {i+1}--")
    print(f"The Values are: {b[i]} and {b[i+1]}")
    if b[i]>b[i+1]:
      print(f"{b[i]} is greater than {b[i+1]}")
      print("Swap")
      print(f"Now the List is: {b}!\n\n")
      b[i],b[i+1]=b[i+1],b[i]
    else:
      print(f"Don't Swap! Go to comparision-{i+2}")
      print(f"Now the List is: {b}\n\n")

print("The Sorted List is:",b)


print("\n\n\n\n")




```

```python
def insertion(ins):
  for i in range(1,len(ins)):
    print(f"---Comparison---{i}")
    current_element=ins[i]
    print(f"Current Element {current_element}")
    pos=i
    while current_element<ins[pos-1] and pos>0:
      print(f"{current_element} is less than {ins[pos-1]}.\n Swap! ")
      ins[pos]=ins[pos-1]
      pos=pos-1
      print(f"The list is now: {ins}")
      
      
    ins[pos]=current_element
    print(f"{current_element} is not less than {ins[pos-1]} ")
    print(f"so the list remains same: {ins}\n")
    




ins=np.array([7,2,9,1,5])

print("::Insertion Sort:: \n")

print("The Unsorted List is:",ins)
print("\n")
insertion(ins)
print("The Sorted List is:",ins)


print("\n\n\n\n")
```
